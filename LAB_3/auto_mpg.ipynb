```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('auto-mpg.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
for column in df.columns:
    print(f"\nUnique values in {column}: {df[column].unique()}")
```

```python
df.hist(figsize=(12, 8))
plt.suptitle("Histograms of Iris Dataset Features", fontsize=16) 
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='cylinders', y='displacement', data=df, hue='mpg')
plt.title('cylinders vs displacement')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='weight', y='acceleration', data=df, hue='mpg')
plt.title('weight vs acceleration')
```

```python
plt.figure(figsize=(8,11))
sns.pairplot(df, hue='mpg')
plt.suptitle("Pairplot of Auto-mpg Dataset Features", fontsize=12)
plt.tight_layout()
```
